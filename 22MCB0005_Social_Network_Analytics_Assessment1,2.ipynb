```python
# maximum and minimum in-degree and out-degree in the graph

in_degrees = [G.in_degree(n) for n in G.nodes()]
max_in_degree = max(in_degrees)
min_in_degree = min(in_degrees)

out_degrees = [G.out_degree(n) for n in G.nodes()]
max_out_degree = max(out_degrees)
min_out_degree = min(out_degrees)

print(max_in_degree)
print(min_in_degree)
print(max_out_degree)
print(min_out_degree)
```

```python
# Calculating the PageRank of the graph
pr = nx.pagerank(G)

# Print the PageRank of each node in the graph
for node, score in pr.items():
    print(f"Node {node}: {score}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
import community.community_louvain as community_louvain
```

```python
# Reference: https://github.com/Geometrein/helsinki-city-bikes/blob/main/Helsinki_city_bikes_network_analysis.ipynb

# Load the dataset
G = nx.karate_club_graph()

# Perform community detection using the Louvain algorithm
communities = nx.algorithms.community.modularity_max.greedy_modularity_communities(G)

# Print the detected communities
for i, community in enumerate(communities):
    print(f"Community {i+1}: {list(community)}")

# Visualize the graph with communities
pos = nx.spring_layout(G)
colors = ['r', 'b', 'g', 'c', 'm', 'y', 'k']
plt.figure(figsize=(8, 6))
for i, community in enumerate(communities):
    nx.draw_networkx_nodes(G, pos, nodelist=list(community), node_color=colors[i], label=f"Community {i+1}")
nx.draw_networkx_edges(G, pos)
plt.legend()
plt.show()
```
