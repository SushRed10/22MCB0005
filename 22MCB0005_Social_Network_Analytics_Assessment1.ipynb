```python
# DIGITAL ASSESSMENT 1 
# REG.NO 22MCB0005
import csv
import pandas as pd

#with open('/content/node_edges.csv', 'r') as f:
#    reader = csv.reader(f)
#    edges = list(reader)

df = pd.read_csv('/content/node_edges.csv')

#For Directed Graph
G = nx.from_pandas_edgelist(df, source='source', target='dest', edge_attr='weight', create_using=nx.DiGraph())

nx.draw(G,with_labels=1,font_weight="bold")

G.edges()
G.nodes()
```

```python
#For Undirected Graph
G1 = nx.from_pandas_edgelist(df, source='source', target='dest', edge_attr='weight', create_using=nx.Graph())

nx.draw(G1,with_labels=2,font_weight="bold")
```

```python
#Number of nodes with max and min degree

degrees = [G.degree(n) for n in G.nodes()]
max_degree = max(degrees)
min_degree = min(degrees)
print(max_degree)
print(min_degree)

nx.degree_centrality(G) # Degree Centrality
```

```python
nx.in_degree_centrality(G) # In Degree Centrality
```

```python
nx.out_degree_centrality(G)  # Out Degree Centrality
```

```python
nx.closeness_centrality(G)  # Closeness Centrality
```

```python
nx.betweenness_centrality(G)  #Betweenness Centrality
```

```python
nx.eigenvector_centrality_numpy(G)  # Eigen Vector Centrality
```

```python
# maximum and minimum in-degree and out-degree in the graph

in_degrees = [G.in_degree(n) for n in G.nodes()]
max_in_degree = max(in_degrees)
min_in_degree = min(in_degrees)

out_degrees = [G.out_degree(n) for n in G.nodes()]
max_out_degree = max(out_degrees)
min_out_degree = min(out_degrees)

print(max_in_degree)
print(min_in_degree)
print(max_out_degree)
print(min_out_degree)
```

```python
# Adjacency matrix of the directed and undirected graph
A = nx.to_numpy_array(G)
print(A)
```

```python
# Calculating the PageRank of the graph
pr = nx.pagerank(G)

# Print the PageRank of each node in the graph
for node, score in pr.items():
    print(f"Node {node}: {score}")
```
